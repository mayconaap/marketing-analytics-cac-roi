```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print ("Bibliotecas importadas com sucesso!")
print (f"Pandas versão: {pd.__version__}")
```

```python
# Carregando os dados
df = pd.read_csv('marketing_and_product_performance.csv')

# Ver as primeiras linhas do DataFrame
df.head()
```

```python
# Entendo a estrutura dos dados
print ("Shape do DataFrame:", df.shape)
print ("\nColunas:")
print (df.columns.tolist())
print ("\nTipos de dados:")
print (df.dtypes)
```

```python
# Verificando valores ausentes
df.isnull().sum()
```

```python
# Listar todas as colunas
df.columns.tolist()
```

```python
# Observando as primeiras 5 linhas para entender os dados
df.head()
```

```python
# Ver estatísticas das colunas numéricas
df.describe()
```

```python
# Ver distribuição de conversões e budget
print ("Resumo de Budget e Conversões:")
print (f"Budget médio: R$ {df['Budget'].mean():,.2f}")
print (f"Conversões média: {df['Conversions'].mean():,.0f}")
print (f"ROI médio: {df['ROI'].mean():.1f}%")
```

```python
# Calcular por campanha
df['CAC'] = df['Budget'] / df['Conversions']

# Ver as primeiras campanhas com CAC calculado
df[['Campaign_ID', 'Budget', 'Conversions', 'CAC']].head()
```

```python
# Estatística detalhada do ROI
# O resultado do ROI está muito baixo, então vamos verificar se há campanhas com ROI negativo
print("Estatísticas detalhadas do ROI:")
print(df['ROI'].describe())

# Verificar quantas campanhas tem ROI negativo
roi_negativo = df[df['ROI'] < 0]
print(f"\nCampanhas com ROI negativo: {len(roi_negativo)}")
print(f"Percentual: {len(roi_negativo) / len(df) * 100:.1f}%")
```

```python
# Verificar se já existe coluna de CAC
if 'CAC' in df.columns:
    print("A coluna 'CAC' já existe!")
    print(f"CAC médio: R$ {df['CAC'].mean():,.2f}")
    print(f"CAC mediano: R$ {df['CAC'].median():,.2f}")

else:
    print("X A coluna 'CAC' não existe, calculando agora...")

# Calcular CAC (Custo de Aquisição de Cliente)
df['CAC'] = df['Budget'] / df['Conversions']
print(f"CAC médio: R$ {df['CAC'].mean():,.2f}")
```

```python
# Analisar as campanhas com CAC alto
# Verificar as 10 campanhas com maior CAC
df[['Campaign_ID', 'Budget', 'Conversions', 'CAC', 'ROI']].sort_values('CAC', ascending=False).head(10)
```

```python
# Verificar campanhas com conversão zero
zero_conv = df[df['Conversions'] == 0]
print(f"Campanhas com conversão zero: {len(zero_conv)}")
if len(zero_conv) > 0:
    print("\nEssas campanhas tem CAC infinito:")
    print(zero_conv[['Campaign_ID', 'Budget', 'Conversions', 'ROI']])
    
```

```python
# Comparar ROI médio com essas campanhas ruins
# Remover campanhas com ROI negativo ou conversão zero para calcular ROI médio "saudável"

df_sem_ruins = df[(df['ROI'] >= 0) & (df['Conversions'] > 0)]

print(f"ROI médio geral: {df['ROI'].mean():.1f}%")
print(f"ROI médio sem campanhas ruins: {df_sem_ruins['ROI'].mean():.1f}%")
print(f"\nDiferença: {df['ROI'].mean() - df_sem_ruins['ROI'].mean():.1f}%")
```

```python
# Visualizar ROI com boxplot para identificar outliers
plt.figure(figsize=(10, 5))
plt.boxplot(df['ROI'])
plt.title('Distribuição do ROI das campanhas')
plt.ylabel('ROI (%)')
plt.axhline(y=0, color='r', linestyle='--', alpha =0.5)
plt.show()
```

```python
# Identificar campanhas problemáticas
# Campanhas com CAC muito alto (acima de R$500)
alto_cac = df[df['CAC'] > 500].sort_values('CAC', ascending=False)
print(f"Campanhas com CAC acima de R$500: {len(alto_cac)}")
print("\nTop 10 piores CAC:")
print(alto_cac[['Campaign_ID', 'Budget', 'Conversions', 'CAC', 'ROI']].head(10))
```

```python
# Identificar campanhas de sucesso
# Campanhas com bom ROI e CAC baixo
sucesso = df[(df['ROI'] > 4) & (df['CAC'] < 100)].sort_values('ROI', ascending=False)
print(f"Campanhas de sucesso: {len(sucesso)}")
print("\nTop 10 melhores campanhas:")
print(sucesso[['Campaign_ID', 'Budget', 'Conversions', 'CAC', 'ROI']].head(10))
```

```python
# Análise de retenção
# Analisar tempo de assinatura por cliente
print("Distribuição de tempo de assinatura (meses):")
print(df['Subscription_Length'].describe())

# Relação entre satisfação e tempo de assinatura
satisfacao_tempo = df.groupby('Customer_Satisfaction_Post_Refund')['Subscription_Length'].mean()
print("\nTempo médio de assinatura por nível de satisfação:")
print(satisfacao_tempo)
```

```python
# Gráfico comparativo
# Verificar se temos dados para plotar
if len(sucesso) > 0 and len(problemas) > 0:
    fig, axes = plt.subplots(1, 2, figsize=(12,5))

    # Gráfico 1: ROI
    axes[0].bar(['Sucesso', 'Problemáticas'], 
                [sucesso['ROI'].mean(), problemas['ROI'].mean()],
                color=['green', 'red'])
    axes[0].set_title('ROI Médio por Tipo de Campanha')
    axes[0].set_ylabel('ROI (%)')
    axes[0].axhline(y=0, color='black', linestyle='-', linewidth=0.5)

    # Gráfico 2: CAC
    axes[1].bar(['Sucesso', 'Problemáticas'], 
                [sucesso['CAC'].mean(), problemas['CAC'].mean()],
                color=['green', 'red'])
    axes[1].set_title('CAC Médio por Tipo de Campanha')
    axes[1].set_ylabel('CAC (R$)')

    plt.tight_layout()
    plt.show()
else:
    print("⚠️ Dados insuficientes para gerar gráficos comparativos.")
    print(f"Campanhas de sucesso: {len(sucesso)}")
    print(f"Campanhas problemáticas: {len(problemas)}")
```
